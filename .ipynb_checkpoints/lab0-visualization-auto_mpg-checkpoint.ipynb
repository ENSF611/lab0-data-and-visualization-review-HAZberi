```python
# Import important working libraries
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#Loading the data with na_values flag.
data = data = pd.read_csv('datasets/auto-mpg.csv', names=["mpg", "cylinders", "displacement", "horsepower", "weight","acceleration", "model year", "origin" ,"car name"], na_values='?')
```

```python
#Plotting all the columns
data.plot()
```

```python
#A peek at mpg distribution with a histogram
data['mpg'].plot.hist();
```

```python
#How many cylinders are we dealing with in the data 
data.origin.value_counts()
```

```python
#Do we have similar mpgs for different data origins
axs = data.hist(column='mpg', by='origin')
```

```python
axs = data.hist(column='mpg', by='origin')
axs[0][0].set(title='origin 1', ylim=[0, 45])
axs[0][1].set(title='origin 2', ylim=[0, 45])
axs[1][0].set(title='origin 3', ylim=[0, 45]);
```

```python
#Is mpg and horsepower are correlated.
data.plot.scatter('mpg', 'horsepower', c='origin', colormap='winter');
```

```python
fig, ax = plt.subplots()
data.plot.scatter('mpg', 'horsepower', c='origin', colormap='viridis', ax=ax);
```

```python
#Improving Layout for the above scatter plot
colors = {1: 'tab:blue', 2: 'tab:orange', 3: 'tab:green'}
fig, ax = plt.subplots()
for key, group in data.groupby(by='origin'):
    group.plot.scatter('mpg', 'horsepower', c=colors[key], label=key, ax=ax);
```

```python
import seaborn as sns
```

```python
# Use origin to split mpg into columns
sns.displot(x='mpg', col='origin', data=data);
```

```python
# Use origin to color (hue) in the same plot
sns.displot(x='mpg', hue='origin', data=data);
```

```python
# Use origin to color (hue) and stacked in the same plot 
sns.displot(x='mpg', hue='origin', data=data, multiple='stack');
```

```python
#Box plot for origin and mpg
sns.boxplot(x='origin', y='mpg', data=data);
```

```python
#Making legends
data['origin_str'] = data['origin'].replace([1, 2, 3], ['origin 1', 'origin 2', 'origin 3'])
```

```python
ax = sns.scatterplot(x='mpg', y='horsepower', data=data, hue='origin_str')
```

```python
#Adding regression line
ax = sns.lmplot(x='mpg', y='horsepower', data=data, hue='origin_str')
```

```python
#Pairplot to get an overview of correlations
sns.pairplot(data, vars=['mpg', 'horsepower','weight' , 'cylinders'], hue='origin_str');
```

```python
#View correlations in the data set as a heatmap matrix
g = sns.heatmap(data[['mpg', 'horsepower','weight' , 'cylinders']].corr(method='spearman'), 
            annot=True)
```
